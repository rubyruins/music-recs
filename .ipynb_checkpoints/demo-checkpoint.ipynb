```python
import pandas as pd
import numpy as np
```

```python
triplets = pd.read_table("10000.txt", header=None, names=['user_id', 'song_id', 'listen_count'])
metadata = pd.read_csv("song_data.csv")
```

```python
print(len(metadata))
print(len(triplets))
```

```python
songs = pd.merge(triplets, metadata.drop_duplicates(['song_id']), on='song_id', how='left')
```

```python
len(songs)
```

```python
songs.head(50)
```

```python
def full_name(title, artist):
    return (title + ' - ' + artist)
        
songs['name'] = songs.apply(lambda x: full_name(x['title'], x['artist_name']), axis=1)
```

```python
songs['name']
```

```python
songs_grouped = songs.groupby(['name']).agg({'listen_count': 'count'}).reset_index()
total_listens = songs_grouped['listen_count'].sum()
print(total_listens)
songs_grouped['percentage']  = songs_grouped['listen_count'].div(total_listens)*100
songs_grouped.sort_values(by='percentage', ascending=False)
```
